```python
'''
    Time complexity: O(Q*N)
    For each push operation O(N); O(1) for all other operations.
    
    Space complexity: O(max(N1, N2))

    where Q is the number of queries, N denotes the maximum number of elements in the queue and
    ‘N1' and ‘N2’ denote the size of queues ‘q1’ and ‘q2’.
'''

from sys import stdin, setrecursionlimit
from collections import deque

setrecursionlimit(10 ** 7)


class Stack:
    def __init__(self):
        self.q1 = deque()
        self.q2 = deque()

    def getSize(self):
        
        return len(self.q1)  

    def isEmpty(self):
        return self.getSize() == 0

    def push(self, ele):
        
        self.q2.append(ele)  
    
        while len(self.q1) > 0:  
            self.q2.append(self.q1[0])
            self.q1.popleft()
        
        
        temp = self.q1     
        self.q1 = self.q2
        self.q2 = temp

    def pop(self):
        if self.isEmpty():
            return -1

        
        ans = self.q1[0]   
        self.q1.popleft()
        return ans

    def top(self):
        if self.isEmpty():
            return -1

        ans = self.q1[0]
        return ans





def takeInput():
    values = list(map(int, stdin.readline().strip().split(" ")))
    return values


#  main
st = Stack()
queries = int(input().strip())
for _ in range(queries):
    values = takeInput()
    if values[0] == 1:
        st.push(values[1])

    elif values[0] == 2:
        print(st.pop())

    elif values[0] == 3:
        print(st.top())

    elif values[0] == 4:
        print(st.getSize())

    else:
        print(str(st.isEmpty()).lower())
```
